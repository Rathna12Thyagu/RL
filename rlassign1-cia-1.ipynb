```python
import pandas as pd
import numpy as np
import random
```

```python
movies = pd.read_csv(r'/kaggle/input/rlrecommendationsystem/movies.csv')
movies.head()
```

```python
movies.shape
```

```python
epsilon = 0.1 
alpha = 0.1 
decay_rate = 0.99  

# Initialize Q-values
users = [100,101,102,103,104,105,106,107,108,109]
users = np.array(users)
movie_id = movies['movieId'].unique()
movie_id_list = movie_id[:10]
title = movies['title'].unique()[:10]
Q_values = {user: {movie: 0 for movie in movie_id_list} for user in users}
```

```python
d = {'User_Id':users}
newdf = pd.DataFrame(data=d)
findf = newdf.merge(movies[:10],how='cross')
```

```python
findf['rating'] = 0
```

```python
rate_choice = [0,1]
rating = []
for i in range(len(findf)):
    rating.append(random.choice(rate_choice))
```

```python
findf['rating'] = rating
```

```python
findf
```

```python
def recommend_movie(user_id):
    if random.random() < epsilon:
        return random.choice(movie_id_list) #explore
    else:
        return max(Q_values[user_id], key=Q_values[user_id].get) #exploit
```

```python
# Simulate the training process
for index, row in findf.iterrows():
    user_id = row['User_Id']
    movie_id = row['movieId']

    recommended_movie = recommend_movie(user_id) #recommend a movie
    
    if recommended_movie == movie_id:
        reward = row['rating']
    else:
        reward = 0 

    Q_values[user_id][recommended_movie] += alpha * (reward - Q_values[user_id][recommended_movie])
    epsilon *= decay_rate
```
